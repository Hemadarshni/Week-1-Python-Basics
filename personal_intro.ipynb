```python
print("Welcome to the Personal Introduction Program!\n")

name = input("What is your name? ")
age = input("How old are you? ")
hobby = input("What is your favorite hobby? ")
food = input("What is your favorite food? ")

favorite_things = [hobby, food]

print("\nWelcome", name + "!")
print(f"You are {age} years old and enjoy {favorite_things[0]}.")
print(f"Your favorite food is {favorite_things[1]}.")
print("Thanks for sharing your details!")
```
